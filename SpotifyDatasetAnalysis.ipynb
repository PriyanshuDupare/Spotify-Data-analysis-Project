```python
# Importation of Python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# calculating number of null values in column artists code 

null_count = df['artists'].isnull().sum()
print(null_count)
```

```python
#Filling Null values
df.fillna({'artists': 0}, inplace=True)
```

```python
# calculating number of null values in column album name code 

null_count = df['album_name'].isnull().sum()
print(null_count)
```

```python
#Filling Null values
df.fillna({'album_name': 0}, inplace=True)
```

```python
# calculating number of null values in column album name code 

null_count = df['track_name'].isnull().sum()
print(null_count)
```

```python
#Filling Null values
df.fillna({'track_name': 0}, inplace=True)
```

```python
#Listeners Segmentation
#Group Listeners based on track genre
# Types of Listeners

types_of_genre = df['track_genre'].unique()
print(types_of_genre)
```

```python
# Number of listeners in each segment 

listener_segmentation = df['track_genre'].value_counts().reset_index() 
listener_segmentation = listener_segmentation.rename(columns={'index': 'Genre Type', 'Segment': 'Total Count'})
print(listener_segmentation)
```

```python
#Artists Segmentation
#Group Artists based on track genre
# Types of Artists

types_of_artists = df['artists'].unique()
print(types_of_artists)
```

```python
# Number of Artists in each segment 

artists_segmentation = df['artists'].value_counts().reset_index() 
artists_segmentation = artists_segmentation.rename(columns={'artists': 'Artist Name', 'count': 'Total Count'})
print(artists_segmentation)
```

```python
#Sort top ten most listened artists
repeat_artists_sorted = artists_segmentation.sort_values(by='Total Count', ascending=False)
print(repeat_artists_sorted.head(10))
```

```python
data = {
    'Artist Name': [
        'The Beatles', 'George Jones', 'Stevie Wonder', 'Linkin Park', 
        'Ella Fitzgerald', 'Prateek Kuhad', 'Feid', 'Chuck Berry', 
        'Håkan Hellström', 'OneRepublic'
    ],
    'Total Count': [279, 271, 236, 224, 222, 217, 202, 190, 183, 181]
}
repeat_artists_sorted = pd.DataFrame(data)

# Convert 'Artist Name' to string if it's not already
repeat_artists_sorted['Artist Name'] = repeat_artists_sorted['Artist Name'].astype(str)

# Plotting a horizontal bar graph
plt.barh(repeat_artists_sorted['Artist Name'], repeat_artists_sorted['Total Count'])

# Invert the y-axis to have the highest count at the top
plt.gca().invert_yaxis() #These helps to get the artist name which have highest count on first place

# Labels
plt.title('Top Ten Most Listened Artists')
plt.xlabel('Total Count')
plt.ylabel('Artist Name')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Most listened in each segment 

album_segmentation = df['album_name'].value_counts().reset_index() 
album_segmentation = album_segmentation.rename(columns={'album_name': 'Album Name', 'count': 'Total Count'})
print(album_segmentation)
```

```python
#Sort Most 5 Listened albums
from matplotlib import style
repeat_album_sorted = album_segmentation.sort_values(by='Total Count', ascending=False)
print(repeat_album_sorted.head(5))
```

```python
#plotting a pie chart for these #used when the characters are not in string and are in object
data = {
    'Album Name': [
        'Alternative Christmas 2022', 'Feliz Cumpleaños con Perreo', 
        'Metal', 'Halloween con perreito', 'Halloween Party 2022'
    ],
    'Total Count': [195, 184, 143, 123, 115]
}
repeat_album_sorted = pd.DataFrame(data)

# Get the top 5 albums
top_five_albums = repeat_album_sorted.head(5)

# Explode the album with the highest count
explode = [0.2,0,0,0,0]

# Plotting a pie chart
#plt.figure(figsize=(10, 7))
plt.pie(top_five_albums['Total Count'], labels=top_five_albums['Album Name'], explode=explode, autopct='%1.1f%%', startangle=90)

# Title
plt.title('Top 5 Most Listened Albums')
style.use("dark_background")

# Display the plot
plt.show()
```

```python
# track name per danceability score

# Group the data by track name and calculate the total danceability scores for each track
dance_sales = df.groupby(['track_name'])['danceability'].sum().reset_index()

# Sort the track songs based on their total score in descending order to identify top dance song
top_dance_song = dance_sales.sort_values(by='danceability', ascending=False)

# Print the states 
print(top_dance_song.head(5).reset_index(drop=True))
```

```python
data = {
    'track_name': [
        'Run Rudolph Run', 'Last Last', 'CÓMO SE SIENTE - Remix', 'Frosty The Snowman',
        'X ÚLTIMA VEZ'
    ],
    'danceability': [97.871, 59.625, 51.648, 48.021, 47.096]
}
top_dance_song = pd.DataFrame(data)

# Plotting the bar graph
#plt.figure(figsize=(12, 8))
plt.plot(top_dance_song['track_name'], top_dance_song['danceability'],marker='o', linestyle='--')

# Invert y-axis to have the highest score at the top
#plt.gca().invert_yaxis()

# Labels and title
plt.xlabel('Danceability Score')
plt.ylabel('Track Name')
plt.title('Top 10 Dance Tracks by Danceability Score')
plt.xticks(rotation=50)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# track name per popular score

# Group the data by track name and calculate the most popular scores for each track
popular_sales = df.groupby(['track_name'])['popularity'].sum().reset_index()

# Sort the track songs based on their total score popular in descending order to identify top dance song
top_popular_song = popular_sales.sort_values(by='popularity', ascending=False)

# Print the states 
print(top_popular_song.head(5).reset_index(drop=True))
```

```python
data = {
    'track_name': ['Alone', 'Stay', 'Broken', 'Numb', 'Paradise'],
    'popularity': [1618, 1385, 1344, 1341, 1251]
}
top_popular_song = pd.DataFrame(data)

# Plotting the pie chart
#plt.figure(figsize=(10, 7)) these tell us how much size we want of the graph
explode = [0.2,0,0,0,0]# Explode the track with highest popularity

plt.pie(top_popular_song['popularity'], labels=top_popular_song['track_name'], explode=explode, autopct='%1.1f%%', startangle=180)

# Title
plt.title('Top 5 Popular Tracks by Popularity Score')

# Display the plot
plt.show()
```

```python
# track name per energy score

# Group the data by track name and calculate the most popular scores for each track
top_energy = df.groupby(['track_name'])['energy'].sum().reset_index()

# Sort the track songs based on their total score popular in descending order to identify top dance song
top_energy_song = top_energy.sort_values(by='energy', ascending=False)

# Print the states 
print(top_energy_song.head(10).reset_index(drop=True))
```

```python
data = {
    'track_name': [
        'Run Rudolph Run', 'Halloween', 'RUMBATÓN', 'X ÚLTIMA VEZ', 'Ley Seca',
        'Christmas All Over Again', 'Hot in It', 'Last Last', 
        'Little Saint Nick - 1991 Remix', 'Frosty The Snowman'
    ],
    'energy': [131.715, 80.588, 54.540, 49.590, 45.520, 43.226, 42.891, 42.375, 42.028, 41.702]
}
top_energy_song = pd.DataFrame(data)

# Plotting the horizontal bar graph
plt.figure(figsize=(7, 5))
plt.barh(top_energy_song['track_name'], top_energy_song['energy'], color='skyblue')

# Invert y-axis to have the highest score at the top
plt.gca().invert_yaxis()

# Labels and title
plt.xlabel('Energy Score')
plt.ylabel('Track Name')
plt.title('Top 10 Tracks by Energy Score')

# Display the plot
plt.tight_layout()
plt.show()
```
